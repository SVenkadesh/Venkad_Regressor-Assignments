```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
dataset
```

```python
dataset.columns
```

```python
input=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
input
```

```python
output=dataset[['charges']]
output
```

```python
from sklearn.model_selection import train_test_split
x_train,y_train,x_test,y_test=train_test_split(input,output,test_size=0.30,random_state=0)
```

```python
# Model Creation or Learning Phase
# Linear Regression
# the Linear Regression formula Y=wX+b there find W & B except X value 

from sklearn.linear_model import LinearRegression
Regressor=LinearRegression()
Regressor.fit(x_train,y_train)
```
